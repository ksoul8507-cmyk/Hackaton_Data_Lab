## 2. Chargement du jeu de données propre

```python
path_procesesed = os.path.join('..', 'data', 'processed')
```

```python
education_unemployed = pd.read_csv("../data/Data_africa_education_unemployed.csv")
education_inactive = pd.read_csv("../data/Data_africa_education_inactive.csv")
rural_urban = pd.read_csv("../data/Data_employed_rural_urban.csv")
```

## 3. Application des filtres et structuration des colonnes

Application des filtres d'âge et de pays définis précédemment pour obtenir notre jeu de données nettoyé. Nous procédons ensuite au renommage des colonnes pour rendre les variables plus explicites et prêtes pour l'analyse.

```python
civ_unemployed = education_unemployed[
    education_unemployed["ccode"] == "CIV"
]
def filter_data(df, filters):
    df = df.copy()
    
    if "age_group" in filters:
        df = df[df["age_group"].isin(filters["age_group"])]
    
    if "country" in filters:
        df = df[df["ccode"] == filters["country"]]
    
    return df
```

```python
filters = {
    "age_group": ["15-24", "25-34"],
    "country": "CIV"
}

civ_unemployed_clean = filter_data(civ_unemployed, filters)
```

```python
civ_unemployed_clean = civ_unemployed_clean.rename(columns={
    "edu_ilo": "education_level",
    "share_unemp": "unemployment_rate",
    "share_inact": "inactivity_rate"
})
```

## 4. Exportation du jeu de données final

Sauvegarde du DataFrame nettoyé et filtré sous le nom de `civ_unemployed_clean.csv` dans le dossier des données traitées (`../data/processed/`). Ce fichier servira de base pour notre analyse du chômage.

```python
unemployed_path = os.path.join(
    path_procesesed, 'ci_education_unemployed'
)
```

```python
civ_unemployed_clean.to_csv(unemployed_path)
```
