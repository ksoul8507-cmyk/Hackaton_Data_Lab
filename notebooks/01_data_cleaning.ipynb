## 1. Initialisation et Centralisation des Données

Dans cette première étape, nous importons les bibliothèques indispensables (`pandas`, `numpy`, `matplotlib`) et chargeons le fichier principal contenant les indicateurs sur l'éducation et le marché de l'emploi en Afrique. Afin d'optimiser le processus de nettoyage, nous centralisons nos trois tables de données (`unemployed`, `inactive`, et `urban_rural`) au sein d'un dictionnaire Python nommé `dataset`.Cela nous permettra d'inspecter et de traiter nos structures de données de manière automatisée à l'aide de boucles.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno  

```

```python
education_unemployed = pd.read_csv("../data/Data_africa_education_unemployed.csv")
education_inactive = pd.read_csv("../data/Data_africa_education_inactive.csv")
rural_urban = pd.read_csv("../data/Data_employed_rural_urban.csv")

```

## 2. Exploration Visuelle et Statistique Initiale (Head & Describe)

Nous exécutons une première boucle itérative sur notre dictionnaire de données afin d'afficher pour chaque table:
1. **Un aperçu des premières lignes (`.head()`)** : pour observer la structure des colonnes, le format des variables de classification (pays `ccode`, année, niveau d'études `edu_ilo`, tranches d'âge, genre) et les premiers enregistrements.
2. [cite_start]**Un résumé statistique des variables numériques (`.describe()`)** : pour analyser la distribution, les moyennes, ainsi que les valeurs minimales et maximales afin de détecter d'éventuelles anomalies ou valeurs aberrantes (comme des taux plafonnés ou des volumes de population disproportionnés).

```python
dataset ={
    'unemployed' : education_unemployed,
    'inactive' : education_inactive,
    'urban': rural_urban
}
```

```python
for name, df in dataset.items():
    print("Analyse de :", name)
    print("\n notre head est \n", df.head())
    print("\n notre describe est\n",df.describe())
```

## 3. Contrôle des Types de Données et des Valeurs Manquantes (`.info()`)

Pour finaliser cette première phase de diagnostic, nous passons en revue les métadonnées de nos trois tables (`.info()`). Cette inspection technique automatisée est essentielle pour valider deux points critiques avant d'entamer les jointures ou le filtrage:
***L'exhaustivité des données** : nous vérifions le nombre de lignes non nulles pour chaque colonne afin de s'assurer qu'aucun enregistrement essentiel n'est manquant (les trois tables affichent une absence totale de valeurs nulles sur leurs entrées respectives).
***La conformité des types (`Dtype`)** : nous contrôlons la cohérence des formats système (les variables temporelles en `int64`,  les taux ou volumes de population en `float64` et les facteurs démographiques ou géographiques sous forme de chaînes de caractères `str` / `object`)

```python
for name, df in dataset.items():
    print("\nnotre info est :",df.info())
```
