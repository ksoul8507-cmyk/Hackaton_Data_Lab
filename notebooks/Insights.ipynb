# Axe C : Analyse approfondie des disparités de genre sur le marché du travail en Côte d’Ivoire

---

#  Introduction & Question de recherche

L’emploi des jeunes représente aujourd’hui un enjeu majeur pour le développement économique et social de nombreux pays africains. Malgré l’amélioration progressive de l’accès à l’éducation et à la formation, des inégalités importantes persistent sur le marché du travail, notamment entre les hommes et les femmes.

En Côte d’Ivoire, les jeunes âgés de 15 à 35 ans rencontrent des difficultés croissantes d’insertion professionnelle. Ces difficultés semblent encore plus marquées chez les femmes, même pour les niveaux d’éducation élevés.

Dans le cadre de l’**Axe C : Analyse approfondie des disparités de genre**, cette étude vise à mieux comprendre les écarts entre hommes et femmes concernant :
- le chômage,
- l’inactivité,
- et l’influence du niveau d’éducation sur l’accès à l’emploi.

###  Question de recherche

> Comment le genre, l’âge et le niveau d’éducation influencent-ils les disparités d’accès à l’emploi chez les jeunes de 15 à 35 ans en Côte d’Ivoire ?

---

#  Données & Prétraitement

Les données utilisées proviennent de datasets relatifs :
- au chômage selon le niveau d’éducation,
- à l’inactivité,
- et aux caractéristiques démographiques.

Le travail de préparation des données a consisté à :
- filtrer les données pour la Côte d’Ivoire (CIV),
- sélectionner la tranche d’âge 15–35 ans,
- nettoyer les valeurs inutiles,
- renommer les colonnes pour améliorer la lisibilité,
- vérifier les types de données,
- et préparer un dataset propre pour l’analyse.

Les principales variables étudiées sont :
- le genre,
- l’âge,
- le niveau d’éducation,
- le taux de chômage,
- et le taux d’inactivité.

---

#  Méthodologie / Approche analytique

L’analyse repose principalement sur une approche exploratoire et statistique.

Les principales méthodes utilisées sont :
- analyse descriptive,
- comparaison des moyennes,
- analyse des corrélations,
- regroupements par catégories (genre, âge, niveau d’éducation),
- et visualisation des données à l’aide de graphiques.

Cette approche permet d’identifier :
- les groupes les plus vulnérables,
- les écarts entre hommes et femmes,
- ainsi que les tendances liées à l’éducation et à l’âge.

---

#  Présentation des résultats clés

## 1. Les femmes sont plus touchées par le chômage

Les résultats montrent que les femmes présentent des taux de chômage plus élevés que les hommes dans la majorité des catégories étudiées.

---

## 2. Les jeunes de 15 à 35 ans représentent le groupe le plus vulnérable

Les jeunes actifs âgés de 15 à 35 ans affichent des niveaux de chômage particulièrement élevés, notamment chez les femmes.

---

## 3. L’écart de genre augmente avec le niveau d’éducation

Contrairement aux attentes, les femmes diplômées du supérieur présentent des taux de chômage beaucoup plus élevés que les hommes ayant le même niveau d’éducation.

Le niveau tertiaire présente l’écart le plus important.

---

## 4. Existence d’un déséquilibre structurel

Les résultats suggèrent un décalage entre les qualifications obtenues et les opportunités réellement disponibles sur le marché du travail.

Les jeunes femmes diplômées semblent rencontrer davantage de difficultés d’insertion professionnelle.

---

## 5. Chômage et inactivité sont liés

Une corrélation positive modérée a été observée entre le chômage et l’inactivité.

Cela signifie qu’une hausse du chômage peut être accompagnée d’une augmentation de l’inactivité.

---

#  Enseignements & Implications pour les politiques publiques

Cette étude met en évidence plusieurs enjeux importants :
- la persistance des inégalités de genre,
- les difficultés d’insertion des jeunes,
- et les limites du marché du travail à absorber certains profils qualifiés.

Les résultats montrent également que l’éducation seule ne garantit pas une insertion professionnelle équitable, notamment pour les femmes diplômées.

Ces constats soulignent l’importance :
- d’améliorer l’accès à l’emploi des jeunes,
- de réduire les discriminations liées au genre,
- et d’adapter davantage les formations aux besoins réels du marché du travail.

---

#  Recommandations & Perspectives

##  Recommandations

- Renforcer les politiques d’insertion professionnelle des jeunes.
- Développer des programmes spécifiques d’accompagnement des jeunes femmes diplômées.
- Encourager les partenariats entre établissements de formation et entreprises.
- Réduire les inégalités d’accès aux opportunités professionnelles.
- Mettre en place des politiques favorisant l’emploi des jeunes femmes.

---

##  Perspectives

Pour approfondir cette étude, il serait intéressant :
- d’intégrer davantage de variables socio-économiques,
- d’étudier les différences régionales,
- d’utiliser des modèles prédictifs,
- et d’analyser l’évolution future des disparités de genre sur le marché du travail.
