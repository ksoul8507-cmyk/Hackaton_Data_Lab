## 1. Importation des Bibliothèques et Chargement des Données

Cette section sert à configurer notre environnement de travail en important les librairies fondamentales (`pandas`, `numpy`, `matplotlib`). Nous chargeons le jeu de données épuré de la Côte d'Ivoire afin de générer des indicateurs macroéconomiques précis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import os
```

```python
path_procesesed = os.path.join('..', 'data', 'processed')
civ_unemployed_clean = pd.read_csv(os.path.join(path_procesesed,'ci_education_unemployed'))
```

## 2. Analyse Comparative du Chômage et de l'Inactivité selon le Genre

Ici, nous calculons et comparons les moyennes globales du taux de chômage (`unemployment_rate`) et du taux d'inactivité (`inactivity_rate`) entre les hommes et les femmes. Cela permet de quantifier de manière brute les disparités de genre sur l'ensemble de la population étudiée.

```python
civ_unemployed_clean.groupby("gender")["unemployment_rate"].mean()
```

## 7. Matrice de Corrélation entre le Chômage et l'Inactivité

Cette cellule calcule la matrice de corrélation de Pearson entre le taux de chômage (`unemployment_rate`) et le taux d'inactivité (`inactivity_rate`) 

### Interprétation du résultat :
* Le coefficient de corrélation obtenu est d'environ **0,27**.
* Cela indique une **corrélation positive mais faible** entre les deux variables. 
* Bien que les deux indicateurs évoluent globalement dans le même sens (une hausse du chômage s'accompagne d'une légère hausse de l'inactivité), leur relation linéaire n'est pas automatique. Cela montre que l'inactivité et le chômage répondent à des dynamiques distinctes sur le marché du travail en Côte d'Ivoire.

```python
civ_unemployed_clean[[
    "unemployment_rate",
    "inactivity_rate"
]].corr()
```

## Analyse de l'inactivité économique

Les femmes présentent un taux d’inactivité plus élevé que les hommes, indiquant une exclusion plus forte du marché du travail.

```python
civ_unemployed_clean.groupby("gender")["inactivity_rate"].mean()
```

## Impact du niveau d'éducation

Contrairement aux idées reçues, les résultats montrent que le taux de chômage augmente avec le niveau d’éducation.

Les individus ayant un niveau d’éducation supérieur présentent des taux de chômage plus élevés que ceux ayant un niveau d’éducation faible.

Cela peut s’expliquer par :
- des attentes plus élevées des diplômés sur le marché du travail,
- un décalage entre formation et opportunités d’emploi,
- une saturation du marché des diplômés.

```python
civ_unemployed_clean.groupby("education_level")["unemployment_rate"].mean().sort_values()
```

## Interaction genre et éducation

Même à niveau d’éducation équivalent, les femmes restent plus exposées au chômage.

```python
civ_unemployed_clean.groupby(
    ["education_level", "gender"]
)["unemployment_rate"].mean()
```

##  Groupes les plus vulnérables

L’analyse montre un écart de genre systématique en faveur des hommes à tous les niveaux d’éducation.

Cependant, cet écart se creuse fortement avec l’augmentation du niveau d’éducation.

Contrairement à l’intuition classique, les femmes diplômées du niveau tertiaire présentent les taux de chômage les plus élevés, largement supérieurs à ceux des hommes du même niveau.

Cela suggère que l’éducation ne réduit pas automatiquement les inégalités de genre sur le marché du travail.

```python
civ_unemployed_clean.groupby(
    ["gender", "age_group", "education_level"]
)["unemployment_rate"].mean() \
.sort_values(ascending=False)
```
